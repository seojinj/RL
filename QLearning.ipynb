```python
import numpy as np
from msdm.domains import GridWorld
from msdm.core.problemclasses.mdp import TabularPolicy,TabularMarkovDecisionProcess
from msdm.core.assignment import AssignmentMap
import itertools
```

```python
defaultparams = dict(
    initial_features=('s',),
    absorbing_features=('g',),
    wall_features=('#',),
    default_features=('.',),
    feature_rewards={
        'g': 0,
        'x': -20
    },
    step_cost=-1,
)

gw = GridWorld(
    tile_array=[
        '..#..g',
        '..#...',
        '..#...',
        's.....',
    ],
    **defaultparams
)

mdp = gw
S = mdp.state_list # set of all states from MDP
A = mdp.action_list # set of all actions from MDP
P = mdp.transitionmatrix# state transition function
# P is sparse: vector representation > huge (think of as function)
R = mdp.rewardmatrix# reward function
```

```python
stepsize = 0.1
d = 0.99
Q = AssignmentMap()

epsilon = 0.9
policy = epGreedy(Q, epsilon, S, A)
count = 100 #initialize count
Q[S[1]] = AssignmentMap([(a,0) for a in A])
traj = []

for i in range(count):
    # can be same state for each episode
    s = S[1]
    for j in itertools.count():
        
        a_prob = policy(s)
        # print(a_prob)
        a_i = np.random.choice(np.arange( 
                      len(a_prob)), 
                       p = a_prob) 
        
        s_i = S.index(s)
        n_i = np.argmax(P[s_i][a_i])
        n = S[n_i]
        #print(P[s_i][a_i])
        traj.append(s)
        print(i,j,s, a_i, n_i)
        if not Q.__contains__(n):
            Q[n] = AssignmentMap([(a,0) for a in A])
        
        max_q = -float("inf")
        for k,v in AssignmentMap.items(Q[n]):
            if v > max_q:
                max_q = v
        
        a = A[a_i]
        Q[s][a] += stepsize * (R[s_i][a_i][n_i] + d * max_q - Q[s][a])
        
        s = n
        
        if mdp.is_terminal(s):
            break
```

```python
gw.plot().plot_trajectory(traj)
```
